```python
article_list = []
for i, line in df.iterrows():
    a = line['article'].strip()
    str_find = a.find(" ▸ ")
    str_find2 = a.find(" 延伸閱讀\n\r")
    if str_find != -1: 
        a = a[:str_find]
    if str_find2 != -1:
        a = a[:str_find2]


    article_list.append(a)
#     print(a)
#     print(str_find)
#     break
```

```python
article_list[20]
```

```python
df['article_clean'] = article_list
```
